```python
#Import the necessary libraries
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
#Import the dataset
iris_dataset = pd.read_csv(r'C:\Users\ARPITA\Documents\ISHITA\DS Projects\Iris classification\Iris.csv')
```

```python
#Inspect the dataset
iris_dataset.describe()
```

```python
#Seperate the features and target variable
X = iris_dataset.iloc[:, 1:-1].values
y = iris_dataset.iloc[:, -1].values
```

```python
np.unique(y)
```

```python
#Encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
#Splitting data into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#Training decision tree model on the data
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#Predict the results for the test set
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Check model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
#Visualizing training set results
from sklearn import tree
fig = plt.figure(figsize=(25,20))
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
graph = tree.plot_tree(classifier, feature_names=fn, class_names=cn, filled=True)
```
